```python
import seaborn as sns
import matplotlib.pyplot as plt

grouped_data=data.groupby(["SEX","EDUCATION","MARRIAGE"]).size().reset_index()

sns.barplot(x="SEX", y="EDUCATION", hue="MARRIAGE", data=grouped_data)
plt.xlabel('SEX')
plt.ylabel('EDUCATION')
plt.title('Count of each distribution by gender')
plt.show()
```
